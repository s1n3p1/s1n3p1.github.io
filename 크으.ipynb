# 전처리

```python
texts = []

for n in range(1, 8):
    f = open('Talk_2023.8.29 09_07-{}.txt'.format(n), 'r')
    while True:
        line = f.readline()
        if not line: break
        texts.append(line)
    f.close()
```

```python
len(texts)
```

```python
for t in texts:
    if '초대했습니다.' in t or '나갔습니다.' in t:
        print(t)
```

```python
texts = [t for t in texts if t[0] == '2' and t[4] == '.' and '초대했습니다.' not in t and '나갔습니다.' not in t and '오' in t]
len(texts)
```

```python
texts_date = []
for t in texts:
    i = t.find(',')
    texts_date.append((t[:i], t[(i+2):]))
```

```python
texts_date_name = []
for t in texts_date:
    i = t[1].find(':')
    texts_date_name.append([t[0], t[1][:(i-1)], t[1][(i+2):-1]])
```

```python
texts_date_name[:10]
```

```python
# 장호진

for i, t in enumerate(texts_date_name):
    if t[1] == '(알 수 없음)':
        texts_date_name[i][1] = '장호진'
```

# 테이블화

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_row', 10)
```

```python
df = pd.DataFrame(texts_date_name, columns = ['time', 'name', 'text'])
df
```

# 함수 제작

```python
def word_count(word):
    pd.set_option('display.max_row', 20)
    df_w = df[df.text.str.contains(word)]
    df_w = df_w[~df_w.name.str.contains('알 수 없음')]
    df_c = pd.DataFrame(df_w.name.value_counts())
    df_c.columns = ['횟수']
    print("'{}' 언급 횟수".format(word))
    print('===========================')
    return(df_c)
```
